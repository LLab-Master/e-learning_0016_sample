# 機械学習におけるデータの重要性
## 概要

　機械学習とその中でのデータの重要性について学びます。


## 機械学習とは

### AIとは

- 人間の知能を置き換える機械
- 厳密な定義はありません

### 今のAIブーム

- 特定領域に限定したAI(弱いAI、専用AI)
- ディープラーニングの急速な進展による第3次AIブーム
- 実用化されている分野については、人間以上の機能(ひたすら黙々と動くという意味で)

  - 画像認識
  - 音声認識
  - 検索エンジン etc.


## AI/機械学習/ディープラーニング

AI/機械学習/ディープラーニングの関係

- 「AI」人工知能全般
- 「機械学習」学習により自動で改善するようにつくられたコンピュータアルゴリズム
- 「ディープラーニング」"ディープニューラルネットワーク"を用いた「機械学習」の一分野

## 機械学習とデータ

機械学習のキーポイントはデータです。

- 機械学習 = モデル(手法) × データ
- 発展途中とはいえ、すでに実用可能なモデル(手法)は存在します。
- データは対象の問題に応じたものを用意する必要があります。
- データの量が少ないと学習が進みません。
- データの質が悪いと学習の効率や結果の精度が悪くなります


## 前処理の必要性

　集めたデータをそのまま使えるわけではありません。
 
- ディープラーニング以前の機械学習
  - データの特徴を表す指標(特徴量)を人間が考える必要がありました。
  
    => 特徴量がどれだけ的確かが学習の良否を決めます。
  - 加えて教師あり学習では「正解データ」を人間が作る必要ありました。
- ディープラーニング
  - 特徴量自体をコンピュータが抽出します。
  
    => データそのものが特徴を的確に表現していなければいけません。
  - 「正解データ」は依然として人間が用意する必要があります。
